### Attempting to use a MLP

```python
import torch
import torch.nn as nn
import numpy as np
import scipy.io 
import random
import math
import matplotlib.pyplot as plt
import torch.nn.functional as F
import os
import seaborn as sn
import pandas as pd
os.environ['KMP_DUPLICATE_LIB_OK']='True' 

from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader
from sklearn.metrics import accuracy_score
```

```python
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        
        self.layer1 = nn.Linear(54, 64)
        self.bn1 = nn.BatchNorm1d(64)
        
        self.layer2 = nn.Linear(64, 56)
        self.bn2 = nn.BatchNorm1d(56)
        
        self.layer3 = nn.Linear(56, 48)
        self.bn3 = nn.BatchNorm1d(48)
        
        self.layer4 = nn.Linear(48, 40)
        self.bn4 = nn.BatchNorm1d(40)
        
        self.layer5 = nn.Linear(40, 36)
        self.bn5 = nn.BatchNorm1d(36)
        
        self.layer6 = nn.Linear(36, 32)
        self.bn6 = nn.BatchNorm1d(32)
        
        self.layer7 = nn.Linear(32, 28)
        self.bn7 = nn.BatchNorm1d(28)
        
        self.layer8 = nn.Linear(28, 24)
        self.bn8 = nn.BatchNorm1d(24)
        
        self.layer9 = nn.Linear(24, 20)
        self.bn9 = nn.BatchNorm1d(20)
        
        self.layer10 = nn.Linear(20, 16)
        self.bn10 = nn.BatchNorm1d(16)
        
        self.layer11 = nn.Linear(16, 12)
        self.bn11 = nn.BatchNorm1d(12)
        
        self.layer12 = nn.Linear(12, 8)
        self.bn12 = nn.BatchNorm1d(8)
        
        self.layer13 = nn.Linear(8, 4)
        self.bn13 = nn.BatchNorm1d(4)
        
        self.layer14 = nn.Linear(4, 2)

    def forward(self, x):
        x = F.relu(self.bn1(self.layer1(x)))
        x = F.relu(self.bn2(self.layer2(x)))
        x = F.relu(self.bn3(self.layer3(x)))
        x = F.relu(self.bn4(self.layer4(x)))
        x = F.relu(self.bn5(self.layer5(x)))
        x = F.relu(self.bn6(self.layer6(x)))
        x = F.relu(self.bn7(self.layer7(x)))
        x = F.relu(self.bn8(self.layer8(x)))
        x = F.relu(self.bn9(self.layer9(x)))
        x = F.relu(self.bn10(self.layer10(x)))
        x = F.relu(self.bn11(self.layer11(x)))
        x = F.relu(self.bn12(self.layer12(x)))
        x = F.relu(self.bn13(self.layer13(x)))
        x = self.layer14(x)
        return x

net = Net()
```

```python
from sklearn.model_selection import train_test_split
X = nflpast.drop(['TeamPts', 'OppPts'], axis=1)  # Features
y = nflpast[['TeamPts', 'OppPts']]  # Labels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Normalize the data except for column 'Team'
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_numpy = X_train.values
X_test_numpy = X_test.values
X_train_numpy[:,1:] = scaler.fit_transform(X_train_numpy[:,1:])
X_test_numpy[:,1:] = scaler.transform(X_test_numpy[:,1:])
X_train = torch.tensor(X_train_numpy, dtype=torch.float32)
X_test = torch.tensor(X_test_numpy, dtype=torch.float32)
y_train = y_train.values
y_test = y_test.values
```

```python

X_train  = torch.tensor(np.array(X_train), dtype=torch.float32)
X_test = torch.tensor(np.array(X_test), dtype=torch.float32)
y_train = torch.tensor(np.array(y_train), dtype=torch.float32)
y_test = torch.tensor(np.array(y_test), dtype=torch.float32)
print(X_train)
```

```python
#define a loss function and optimizer
from torch.optim.lr_scheduler import StepLR
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(net.parameters(), lr=0.05)
scheduler = StepLR(optimizer, step_size=3, gamma=0.85)
num_epochs = 300  # Define the number of training epochs
for epoch in range(num_epochs):
    optimizer.zero_grad()  # Zero the gradients
    outputs = net(X_train)  # Forward pass
    loss = criterion(outputs, y_train)  # Calculate the loss
    loss.backward()  # Backpropagation
    optimizer.step()  # Update the model parameters
    scheduler.step()
    # Print the loss at each epoch
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}')
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
net.eval()
with torch.no_grad():
    train_predictions = net(X_train)
    test_predictions = net(X_test)
print("Training Predictions:", train_predictions)
print("Training R-squared:", r2_score(y_train, train_predictions))
print("Validation R-squared:", r2_score(y_test, test_predictions))
print("Training MSE:", mean_squared_error(y_train, train_predictions))
print("Validation MSE:", mean_squared_error(y_test, test_predictions))
print("Training MAE:", mean_absolute_error(y_train, train_predictions))
print("Validation MAE:", mean_absolute_error(y_test, test_predictions))
```
