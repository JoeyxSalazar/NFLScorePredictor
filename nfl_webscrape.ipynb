```python
import numpy as np
import pandas as pd
```

```python
seasons = [str(season) for season in range(2023,2023)]
print(f"Number of seasons: {len(seasons)}")
```

```python
team_abb = ['crd', 'atl', 'rav', 'buf','car', 'chi',
            'cin', 'cle', 'dal', 'det', 'gnb', 'htx', 'clt', 'jax', 'kan', 'sdg',
            'ram', 'rai', 'mia', 'nwe', 'nor', 'nyg', 'nyj', 'phi', 'pit', 'sea', 'sfo', 'tam', 'oti', 'was']
```

```python
import random
import time
nfl_df = pd.DataFrame()

for season in seasons:
    for team in team_abb:
        url = 'https://www.pro-football-reference.com/teams/' + team + '/' + season + '/gamelog/'
        print(url)
        off_df = pd.read_html(url, header=1, attrs = {'id': 'gamelog' + season})[0]
        def_df = pd.read_html(url, header=1, attrs = {'id': 'gamelog' + season})[0]
        team_df = pd.concat([off_df, def_df], axis = 1)

        team_df.insert(loc=0, column='Season', value=season)
        team_df.insert(loc=2, column='Team', value=team.upper())
        nfl_df = pd.concat([nfl_df, team_df], ignore_index=True)
        time.sleep(random.randint(5,7))
```
